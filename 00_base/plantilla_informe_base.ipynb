```python
import matplotlib.pyplot as plt

def graficar_bloque(df, df_jugador, bloque_nombre, metricas_bloque):

    # Convertir a numérico
    df[metricas_bloque] = df[metricas_bloque].apply(pd.to_numeric, errors='coerce')

    # Percentiles (CLAVE)
    percentiles = df[metricas_bloque].rank(pct=True)

    # Jugador
    jugador = df_jugador["Player"].iloc[0]
    fila = df[df["Player"] == jugador].index[0]

    datos = percentiles.loc[fila, metricas_bloque].dropna().sort_values()

    # Gráfico
    fig, ax = plt.subplots(figsize=(10, 6))
    bars = ax.barh(datos.index, datos.values)

    # Estilo
    ax.set_title(f"{bloque_nombre} | {jugador}", fontsize=16, pad=15)
    ax.set_xlim(0, 1)

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.grid(axis='x', linestyle='--', alpha=0.25)

    # Color por rendimiento
    for bar in bars:
        if bar.get_width() > 0.75:
            bar.set_color("#4B0082")  # violeta fuerte
        elif bar.get_width() > 0.5:
            bar.set_color("#7B2CBF")
        else:
            bar.set_color("#CDB4DB")

    # Valores
    for i, v in enumerate(datos.values):
        ax.text(v, i, f" {v:.2f}", va='center', fontsize=9)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def graficar_bloque(df, jugador, bloque_nombre, metricas_bloque):
    metricas_existentes = [m for m in metricas_bloque if m in df.columns]

    if len(metricas_existentes) == 0:
        print(f"No hay métricas válidas en el bloque: {bloque_nombre}")
        return

    df_temp = df.copy()
    df_temp[metricas_existentes] = df_temp[metricas_existentes].apply(pd.to_numeric, errors="coerce")

    df_jugador = df_temp[df_temp["Player"] == jugador].copy()

    if df_jugador.empty:
        print(f"No se encontró al jugador: {jugador}")
        return

    percentiles = df_temp[metricas_existentes].rank(pct=True)
    fila = df_jugador.index[0]
    datos = percentiles.loc[fila, metricas_existentes].dropna().sort_values()

    etiquetas = []
    for m in datos.index:
        etiqueta = m.replace(" per 90", "")
        etiqueta = etiqueta.replace(", %", "")
        etiqueta = etiqueta.replace("Accurate ", "")
        etiquetas.append(etiqueta)

    fig, ax = plt.subplots(figsize=(11, 6))
    bars = ax.barh(etiquetas, datos.values)

    ax.set_title(f"{bloque_nombre} | {jugador}", fontsize=18, loc="left", pad=15)
    ax.set_xlim(0, 1)

    ax.set_facecolor("#f2f2f2")
    fig.patch.set_facecolor("#f2f2f2")

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.grid(axis="x", linestyle="--", alpha=0.2)
    ax.tick_params(axis="y", labelsize=10)
    ax.tick_params(axis="x", labelsize=9)

    for bar in bars:
        if bar.get_width() > 0.8:
            bar.set_color("#4B0082")
        elif bar.get_width() > 0.6:
            bar.set_color("#6A0DAD")
        elif bar.get_width() > 0.4:
            bar.set_color("#9D4EDD")
        else:
            bar.set_color("#CDB4DB")

    for i, v in enumerate(datos.values):
        ax.text(v + 0.01, i, f"{v:.2f}", va="center", ha="left", fontsize=9)

    ax.text(1, -1.2, "Javi Mera / @iojavime", ha="right", fontsize=8, alpha=0.6)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def bubble_chart(df, jugador, x, y, size, titulo=None):
    df_temp = df.copy()

    columnas = ["Player", x, y, size]
    df_temp = df_temp[columnas].copy()

    for col in [x, y, size]:
        df_temp[col] = pd.to_numeric(df_temp[col], errors="coerce")

    df_temp = df_temp.dropna(subset=[x, y, size])

    fig, ax = plt.subplots(figsize=(10, 7))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Todos los jugadores
    ax.scatter(
        df_temp[x],
        df_temp[y],
        s=df_temp[size] * 80,
        alpha=0.35
    )

    # Resaltar jugador elegido
    df_j = df_temp[df_temp["Player"] == jugador]
    if not df_j.empty:
        ax.scatter(
            df_j[x],
            df_j[y],
            s=df_j[size] * 120,
            alpha=0.9
        )

        for _, row in df_j.iterrows():
            ax.text(row[x], row[y], row["Player"], fontsize=10)

    ax.set_xlabel(x, fontsize=11)
    ax.set_ylabel(y, fontsize=11)

    if titulo is None:
        titulo = f"{jugador} | {x} vs {y} (size: {size})"

    ax.set_title(titulo, fontsize=15, loc="left", pad=15)

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.grid(alpha=0.2, linestyle="--")
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def bubble_chart_violeta(df, jugador, x, y, size, titulo=None):
    df_temp = df.copy()

    columnas = ["Player", x, y, size]
    df_temp = df_temp[columnas].copy()

    for col in [x, y, size]:
        df_temp[col] = pd.to_numeric(df_temp[col], errors="coerce")

    df_temp = df_temp.dropna(subset=[x, y, size])

    fig, ax = plt.subplots(figsize=(10, 7))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Normalizar tamaños
    size_norm = df_temp[size] / df_temp[size].max()

    # Todos los jugadores (violeta claro)
    ax.scatter(
        df_temp[x],
        df_temp[y],
        s=size_norm * 300,
        color="#CDB4DB",   # violeta claro
        alpha=0.5
    )

    # Jugador destacado (violeta fuerte)
    df_j = df_temp[df_temp["Player"] == jugador]
    if not df_j.empty:
        size_j = (df_j[size] / df_temp[size].max()) * 400

        ax.scatter(
            df_j[x],
            df_j[y],
            s=size_j,
            color="#4B0082",   # violeta oscuro
            alpha=0.9
        )

        for _, row in df_j.iterrows():
            ax.text(row[x], row[y], row["Player"], fontsize=10, ha="left")

    ax.set_xlabel(x, fontsize=11)
    ax.set_ylabel(y, fontsize=11)

    if titulo is None:
        titulo = f"{jugador} | {x} vs {y}"

    ax.set_title(titulo, fontsize=15, loc="left", pad=15)

    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.grid(alpha=0.2, linestyle="--")

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def bubble_informe(df, jugador, x, y, size, titulo, subtitulo):

    df_temp = df.copy()
    columnas = ["Player", x, y, size]
    df_temp = df_temp[columnas].copy()

    for col in [x, y, size]:
        df_temp[col] = pd.to_numeric(df_temp[col], errors="coerce")

    df_temp = df_temp.dropna()

    # Normalizar tamaño
    size_norm = df_temp[size] / df_temp[size].max()

    # Medias
    x_mean = df_temp[x].mean()
    y_mean = df_temp[y].mean()

    fig, ax = plt.subplots(figsize=(12, 8))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Todos los jugadores
    ax.scatter(
        df_temp[x],
        df_temp[y],
        s=size_norm * 800,
        color="#CDB4DB",
        alpha=0.6
    )

    # Jugador destacado
    df_j = df_temp[df_temp["Player"] == jugador]

    if not df_j.empty:
        ax.scatter(
            df_j[x],
            df_j[y],
            s=(df_j[size] / df_temp[size].max()) * 900,
            color="#4B0082",
            edgecolor="black",
            linewidth=1.5,
            alpha=0.95,
            zorder=3
        )

        for _, row in df_j.iterrows():
            ax.text(row[x], row[y], jugador, fontsize=10)

    # Etiquetar TOP jugadores (los más extremos)
    top = df_temp.sort_values(x, ascending=False).head(3)
    for _, row in top.iterrows():
        ax.text(row[x], row[y], row["Player"], fontsize=9)

    # Líneas medias
    ax.axvline(x_mean, linestyle="--", color="gray", alpha=0.6)
    ax.axhline(y_mean, linestyle="--", color="gray", alpha=0.6)

    # Título y subtítulo
    fig.suptitle(titulo, fontsize=20, x=0.125, ha="left")
    ax.set_title(subtitulo, fontsize=11, loc="left", pad=10, alpha=0.7)

    # Ejes
    ax.set_xlabel(x, fontsize=11)
    ax.set_ylabel(y, fontsize=11)

    # Quitar bordes
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Grid
    ax.grid(alpha=0.2, linestyle="--")

    # Firma
    ax.text(
        1, -0.1,
        "Javi Mera / @iojavime",
        transform=ax.transAxes,
        ha="right",
        fontsize=9,
        alpha=0.6
    )

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def grafico_linea_jugadores(df, metrica, volumen=None, jugador_destacado=None, titulo="", subtitulo=""):

    df_temp = df.copy()

    # Convertir a numérico
    df_temp[metrica] = pd.to_numeric(df_temp[metrica], errors="coerce")

    if volumen:
        df_temp[volumen] = pd.to_numeric(df_temp[volumen], errors="coerce")
    else:
        df_temp["volumen_aux"] = 1
        volumen = "volumen_aux"

    df_temp = df_temp.dropna(subset=[metrica, volumen])
    df_temp = df_temp.sort_values(metrica)

    x = df_temp[metrica]
    y = [0]*len(df_temp)

    size_norm = df_temp[volumen] / df_temp[volumen].max()

    fig, ax = plt.subplots(figsize=(12, 3))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Línea base
    ax.hlines(y=0, xmin=x.min(), xmax=x.max(), color="gray", alpha=0.3)

    # Jugadores
    ax.scatter(
        x,
        y,
        s=size_norm * 400,
        color="#CDB4DB",
        alpha=0.7
    )

    # Destacado
    if jugador_destacado:
        df_j = df_temp[df_temp["Player"] == jugador_destacado]

        if not df_j.empty:
            ax.scatter(
                df_j[metrica],
                [0],
                s=(df_j[volumen] / df_temp[volumen].max()) * 500,
                color="#4B0082",
                edgecolor="black",
                linewidth=1.5,
                zorder=3
            )

            for _, row in df_j.iterrows():
                ax.text(row[metrica], 0.02, jugador_destacado, fontsize=10)

    # Extremos
    extremos = pd.concat([df_temp.head(3), df_temp.tail(3)])

    for _, row in extremos.iterrows():
        ax.text(row[metrica], 0.02, row["Player"], fontsize=9, alpha=0.8)

    # TÍTULO Y SUBTÍTULO (CLAVE)
    fig.suptitle(titulo, fontsize=20, x=0.125, ha="left")
    ax.set_title(subtitulo, fontsize=11, loc="left", pad=10, alpha=0.7)

    # Ejes
    ax.set_yticks([])
    ax.set_xlabel(metrica, fontsize=11)

    # Estética
    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.grid(axis="x", linestyle="--", alpha=0.2)

    # Firma
    ax.text(
        1, -0.4,
        "Javi Mera / @iojavime",
        transform=ax.transAxes,
        ha="right",
        fontsize=8,
        alpha=0.6
    )

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def grafico_linea_estilo_ref(
    df,
    metrica,
    jugador_destacado=None,
    titulo="",
    subtitulo="",
    jugadores_extra=None
):
    df_temp = df.copy()
    df_temp[metrica] = pd.to_numeric(df_temp[metrica], errors="coerce")
    df_temp = df_temp.dropna(subset=[metrica]).sort_values(metrica).reset_index(drop=True)

    x = df_temp[metrica]
    y = [0] * len(df_temp)

    fig, ax = plt.subplots(figsize=(16, 5))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Línea base
    ax.hlines(y=0, xmin=x.min(), xmax=x.max(), color="#9a9a9a", alpha=0.6, linewidth=2)

    # Todos los jugadores
    ax.scatter(
        x,
        y,
        s=120,
        color="#b79ad9",
        alpha=0.75,
        edgecolors="none",
        zorder=2
    )

    # Jugador destacado
    if jugador_destacado is not None:
        df_j = df_temp[df_temp["Player"] == jugador_destacado]
        if not df_j.empty:
            xj = df_j[metrica].iloc[0]
            ax.scatter(
                [xj],
                [0],
                s=260,
                color="#7a3db8",
                edgecolor="black",
                linewidth=2.2,
                zorder=5
            )
            ax.text(
                xj,
                0.10,
                jugador_destacado,
                ha="center",
                va="bottom",
                fontsize=17,
                color="#333333"
            )

    # Etiquetas estilo limpio:
    # 1 mínimo a la izquierda arriba
    # 2 máximo a la derecha abajo
    # 3 segundo máximo a la derecha abajo
    if len(df_temp) >= 1:
        row_min = df_temp.iloc[0]
        if row_min["Player"] != jugador_destacado:
            ax.text(
                row_min[metrica],
                0.07,
                row_min["Player"],
                ha="center",
                va="bottom",
                fontsize=13,
                color="#333333"
            )

    if len(df_temp) >= 2:
        row_max = df_temp.iloc[-1]
        if row_max["Player"] != jugador_destacado:
            ax.text(
                row_max[metrica],
                -0.12,
                row_max["Player"],
                ha="center",
                va="top",
                fontsize=13,
                color="#333333"
            )

    if len(df_temp) >= 3:
        row_second_max = df_temp.iloc[-2]
        if row_second_max["Player"] != jugador_destacado:
            ax.text(
                row_second_max[metrica],
                -0.08,
                row_second_max["Player"],
                ha="center",
                va="top",
                fontsize=13,
                color="#333333"
            )

    # Etiquetas extra opcionales
    if jugadores_extra:
        for i, nombre in enumerate(jugadores_extra):
            row = df_temp[df_temp["Player"] == nombre]
            if not row.empty and nombre != jugador_destacado:
                xx = row[metrica].iloc[0]
                yy = -0.08 if i % 2 == 0 else 0.07
                va = "top" if yy < 0 else "bottom"
                ax.text(
                    xx,
                    yy,
                    nombre,
                    ha="center",
                    va=va,
                    fontsize=12,
                    color="#333333"
                )

    # Título y subtítulo bien separados arriba izquierda
    fig.text(0.06, 0.94, titulo, ha="left", va="top", fontsize=30, color="#333333")
    fig.text(0.06, 0.86, subtitulo, ha="left", va="top", fontsize=15, color="#666666")

    # Ejes
    ax.set_yticks([])
    ax.set_xlabel(metrica, fontsize=20, color="#333333")
    ax.tick_params(axis="x", labelsize=18, colors="#333333")

    # Límites verticales para dejar aire a etiquetas
    ax.set_ylim(-0.18, 0.18)

    # Grid suave
    ax.grid(axis="x", linestyle="-", alpha=0.18, color="#9a9a9a")

    # Bordes
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_color("#aaaaaa")

    # Firma
    ax.text(
        1.0,
        -0.16,
        "Javi Mera / @iojavime",
        transform=ax.transAxes,
        ha="right",
        va="top",
        fontsize=16,
        color="#666666"
    )

    # Reservar espacio real arriba para título/subtítulo
    plt.tight_layout(rect=[0.03, 0.05, 0.99, 0.82])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def grafico_linea_estilo_ref(
    df,
    metrica,
    jugador_destacado=None,
    titulo="",
    subtitulo="",
    jugadores_extra=None
):
    df_temp = df.copy()
    df_temp[metrica] = pd.to_numeric(df_temp[metrica], errors="coerce")
    df_temp = df_temp.dropna(subset=[metrica]).sort_values(metrica).reset_index(drop=True)

    x = df_temp[metrica]
    y = [0] * len(df_temp)

    fig, ax = plt.subplots(figsize=(16, 5))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Línea base
    ax.hlines(y=0, xmin=x.min(), xmax=x.max(), color="#9a9a9a", alpha=0.6, linewidth=2)

    # Todos los jugadores
    ax.scatter(
        x,
        y,
        s=120,
        color="#b79ad9",
        alpha=0.75,
        edgecolors="none",
        zorder=2
    )

    # Jugador destacado
    if jugador_destacado is not None:
        df_j = df_temp[df_temp["Player"] == jugador_destacado]
        if not df_j.empty:
            xj = df_j[metrica].iloc[0]
            ax.scatter(
                [xj],
                [0],
                s=260,
                color="#7a3db8",
                edgecolor="black",
                linewidth=2.2,
                zorder=5
            )
            ax.text(
                xj,
                0.10,
                jugador_destacado,
                ha="center",
                va="bottom",
                fontsize=17,
                color="#333333"
            )

    # Etiquetas estilo limpio:
    # 1 mínimo a la izquierda arriba
    # 2 máximo a la derecha abajo
    # 3 segundo máximo a la derecha abajo
    if len(df_temp) >= 1:
        row_min = df_temp.iloc[0]
        if row_min["Player"] != jugador_destacado:
            ax.text(
                row_min[metrica],
                0.07,
                row_min["Player"],
                ha="center",
                va="bottom",
                fontsize=13,
                color="#333333"
            )

    if len(df_temp) >= 2:
        row_max = df_temp.iloc[-1]
        if row_max["Player"] != jugador_destacado:
            ax.text(
                row_max[metrica],
                -0.12,
                row_max["Player"],
                ha="center",
                va="top",
                fontsize=13,
                color="#333333"
            )

    if len(df_temp) >= 3:
        row_second_max = df_temp.iloc[-2]
        if row_second_max["Player"] != jugador_destacado:
            ax.text(
                row_second_max[metrica],
                -0.08,
                row_second_max["Player"],
                ha="center",
                va="top",
                fontsize=13,
                color="#333333"
            )

    # Etiquetas extra opcionales
    if jugadores_extra:
        for i, nombre in enumerate(jugadores_extra):
            row = df_temp[df_temp["Player"] == nombre]
            if not row.empty and nombre != jugador_destacado:
                xx = row[metrica].iloc[0]
                yy = -0.08 if i % 2 == 0 else 0.07
                va = "top" if yy < 0 else "bottom"
                ax.text(
                    xx,
                    yy,
                    nombre,
                    ha="center",
                    va=va,
                    fontsize=12,
                    color="#333333"
                )

    # Título y subtítulo bien separados arriba izquierda
    fig.text(0.06, 0.94, titulo, ha="left", va="top", fontsize=30, color="#333333")
    fig.text(0.06, 0.86, subtitulo, ha="left", va="top", fontsize=15, color="#666666")

    # Ejes
    ax.set_yticks([])
    ax.set_xlabel(metrica, fontsize=20, color="#333333")
    ax.tick_params(axis="x", labelsize=18, colors="#333333")

    # Límites verticales para dejar aire a etiquetas
    ax.set_ylim(-0.18, 0.18)

    # Grid suave
    ax.grid(axis="x", linestyle="-", alpha=0.18, color="#9a9a9a")

    # Bordes
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_color("#aaaaaa")

    # Firma
    ax.text(
        1.0,
        -0.16,
        "Javi Mera / @iojavime",
        transform=ax.transAxes,
        ha="right",
        va="top",
        fontsize=16,
        color="#666666"
    )

    # Reservar espacio real arriba para título/subtítulo
    plt.tight_layout(rect=[0.03, 0.05, 0.99, 0.82])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random

def grafico_linea_elegante(
    df,
    metrica,
    jugador_destacado=None,
    titulo="",
    subtitulo="",
    n_labels=4,
    seed=42
):
    df_temp = df.copy()
    df_temp[metrica] = pd.to_numeric(df_temp[metrica], errors="coerce")
    df_temp = df_temp.dropna(subset=[metrica]).sort_values(metrica).reset_index(drop=True)

    x = df_temp[metrica].values
    y = np.zeros(len(df_temp))

    fig, ax = plt.subplots(figsize=(16, 5))
    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Línea base
    ax.hlines(y=0, xmin=x.min(), xmax=x.max(), color="#a8a8a8", alpha=0.65, linewidth=1.5, zorder=1)

    # Puntos base
    ax.scatter(
        x,
        y,
        s=70,
        color="#bca3de",
        alpha=0.8,
        edgecolors="none",
        zorder=2
    )

    # Jugador destacado
    xj = None
    if jugador_destacado is not None:
        df_j = df_temp[df_temp["Player"] == jugador_destacado]
        if not df_j.empty:
            xj = df_j[metrica].iloc[0]
            ax.scatter(
                [xj],
                [0],
                s=260,
                color="#6d28a8",
                edgecolor="black",
                linewidth=1.8,
                zorder=5
            )

            # etiqueta objetivo con caja redondeada
            ax.annotate(
                jugador_destacado,
                xy=(xj, 0),
                xytext=(xj, 0.085),
                ha="center",
                va="bottom",
                fontsize=10,
                color="#222222",
                bbox=dict(
                    boxstyle="round,pad=0.25,rounding_size=0.8",
                    fc="#f2f2f2",
                    ec="none",
                    alpha=0.95
                ),
                arrowprops=dict(
                    arrowstyle="-",
                    color="#7a7a7a",
                    lw=0.8,
                    shrinkA=4,
                    shrinkB=6
                ),
                zorder=6
            )

    # Selección aleatoria de nombres
    random.seed(seed)
    candidatos = df_temp.copy()

    if jugador_destacado is not None:
        candidatos = candidatos[candidatos["Player"] != jugador_destacado]

    # evitar extremos demasiado pegados si quieres algo más limpio
    if len(candidatos) > n_labels:
        seleccion = candidatos.sample(n=n_labels, random_state=seed).sort_values(metrica)
    else:
        seleccion = candidatos.sort_values(metrica)

    # posiciones alternadas y controladas para no superponer demasiado
    offsets_y = [0.07, -0.075, 0.10, -0.105, 0.055, -0.06, 0.12, -0.12]
    used_x = []

    for i, (_, row) in enumerate(seleccion.iterrows()):
        xx = row[metrica]

        # pequeña lógica para evitar nombres demasiado juntos
        nudged_x = xx
        for prev_x in used_x:
            if abs(nudged_x - prev_x) < 0.025 * (x.max() - x.min()):
                nudged_x += 0.012 * (x.max() - x.min())

        used_x.append(nudged_x)

        yy = offsets_y[i % len(offsets_y)]
        va = "bottom" if yy > 0 else "top"

        ax.annotate(
            row["Player"],
            xy=(xx, 0),
            xytext=(nudged_x, yy),
            ha="center",
            va=va,
            fontsize=8,
            color="#333333",
            arrowprops=dict(
                arrowstyle="-",
                color="#8a8a8a",
                lw=0.7,
                shrinkA=3,
                shrinkB=5
            ),
            zorder=4
        )

    # Título y subtítulo más elegantes
    fig.text(0.06, 0.93, titulo, ha="left", va="top", fontsize=22, color="#333333")
    fig.text(0.06, 0.865, subtitulo, ha="left", va="top", fontsize=11, color="#666666")

    # Ejes
    ax.set_yticks([])
    ax.set_xlabel(metrica, fontsize=13, color="#333333")
    ax.tick_params(axis="x", labelsize=11, colors="#333333")

    # Aire vertical
    ax.set_ylim(-0.16, 0.16)

    # Grid suave
    ax.grid(axis="x", linestyle="-", alpha=0.14, color="#9a9a9a")

    # Bordes
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_color("#aaaaaa")
    ax.spines["bottom"].set_linewidth(0.8)

    # Firma muy pequeña
    ax.text(
        1.0,
        -0.12,
        "Javi Mera / @iojavime",
        transform=ax.transAxes,
        ha="right",
        va="top",
        fontsize=8,
        color="#666666"
    )

    plt.tight_layout(rect=[0.03, 0.06, 0.99, 0.84])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random

def grafico_linea_elegante(
    df,
    metrica,
    jugador_destacado=None,
    titulo="",
    subtitulo="",
    n_labels=4,
    seed=42
):
    df_temp = df.copy()
    df_temp[metrica] = pd.to_numeric(df_temp[metrica], errors="coerce")
    df_temp = df_temp.dropna(subset=[metrica]).sort_values(metrica).reset_index(drop=True)

    x = df_temp[metrica].values
    y = np.zeros(len(df_temp))

    # 🔥 AQUÍ ESTÁ EL CAMBIO CLAVE (PANTALLA GRANDE)
    fig, ax = plt.subplots(figsize=(22, 6))

    fig.patch.set_facecolor("#f2f2f2")
    ax.set_facecolor("#f2f2f2")

    # Línea base
    ax.hlines(y=0, xmin=x.min(), xmax=x.max(), color="#a8a8a8", alpha=0.65, linewidth=1.5)

    # Puntos
    ax.scatter(
        x,
        y,
        s=70,
        color="#bca3de",
        alpha=0.8
    )

    # Destacado
    if jugador_destacado:
        df_j = df_temp[df_temp["Player"] == jugador_destacado]
        if not df_j.empty:
            xj = df_j[metrica].iloc[0]

            ax.scatter(
                [xj],
                [0],
                s=260,
                color="#6d28a8",
                edgecolor="black",
                linewidth=1.8,
                zorder=5
            )

            ax.annotate(
                jugador_destacado,
                xy=(xj, 0),
                xytext=(xj, 0.09),
                ha="center",
                fontsize=10,
                bbox=dict(boxstyle="round,pad=0.25", fc="#f2f2f2", ec="none"),
                arrowprops=dict(arrowstyle="-", color="#7a7a7a", lw=0.8)
            )

    # Labels aleatorios
    random.seed(seed)
    candidatos = df_temp[df_temp["Player"] != jugador_destacado]

    if len(candidatos) > n_labels:
        seleccion = candidatos.sample(n=n_labels, random_state=seed).sort_values(metrica)
    else:
        seleccion = candidatos

    offsets = [0.08, -0.09, 0.11, -0.12]

    for i, (_, row) in enumerate(seleccion.iterrows()):
        xx = row[metrica]
        yy = offsets[i % len(offsets)]

        ax.annotate(
            row["Player"],
            xy=(xx, 0),
            xytext=(xx, yy),
            ha="center",
            fontsize=8,
            arrowprops=dict(arrowstyle="-", color="#8a8a8a", lw=0.7)
        )

    # Títulos (más elegantes)
    fig.text(0.05, 0.94, titulo, fontsize=20)
    fig.text(0.05, 0.89, subtitulo, fontsize=11, alpha=0.7)

    # Ejes
    ax.set_yticks([])
    ax.set_xlabel(metrica, fontsize=12)

    ax.set_ylim(-0.18, 0.18)

    # Estética
    for spine in ax.spines.values():
        spine.set_visible(False)

    ax.grid(axis="x", alpha=0.15)

    # Firma pequeña
    ax.text(
        1, -0.12,
        "Javi Mera / @iojavime",
        transform=ax.transAxes,
        ha="right",
        fontsize=8,
        alpha=0.6
    )

    plt.tight_layout(rect=[0.03, 0.06, 0.99, 0.85])
    plt.show()
```
